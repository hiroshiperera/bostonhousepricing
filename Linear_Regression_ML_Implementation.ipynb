```python
# Independent and Dependent Features
X = boston_df.iloc[:,:-1]
y = boston_df.iloc[:,-1]
```

```python
# Train Test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
# 
```

```python
# Standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)
```

```python
X_test
```

# Model Training

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## on which parameters the model has been trained
regression.get_params()
```

```python
## prediction with the test data

reg_pred = regression.predict(X_test)
```

```python
## plot a scatter plot for the prediction
plt.scatter(y_test, reg_pred)

# This model is linear, which indicates something great
```

```python
## Prediction with reciduals

residuals = y_test-reg_pred
residuals
```

```python
## Plot with residuals

sns.displot(residuals, kind = "kde")
#This is mostly like a normalized dataset
```

```python
## Scatter plot with respect to predictions and residuals
plt.scatter(reg_pred, residuals)
# Residuals are scattered and we cannot see a uniformed distribution 
```

```python
# Performance metric

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
```

```python
# Rsquare and Adjusted R square

from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
print(score)
```

```python
# Display adjusted R-squared

1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
